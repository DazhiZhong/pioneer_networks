```python
def drawgraph(G):
    tree = nx.bfs_tree(G,7)
    q1 = [7]
    q2 = []
    passed = []
    level = 0
    co = {}
    while True:
        while q1!=[]:
            n = q1.pop(0)
            co[n]=level
            for i in nx.neighbors(tree, n):
                q2.append(i)
        if q2!=[]:
            q1 = q2.copy()
            q2 = []
            level+=1
        elif q2==[]:
            break
    print(co)
    colors = []
    for n in G.nodes:
        colors.append(co[n])
    print(colors)
    nx.draw_networkx(G,node_color=colors, cmap=plt.get_cmap('YlOrRd'))
```

```python
def drawgraph_trim(G):
    tree = nx.bfs_tree(G,7)
    q1 = [7]
    q2 = []
    passed = []
    level = 0
    co = {}
    while True:
        while q1!=[]:
            n = q1.pop(0)
            co[n]=level
            for i in nx.neighbors(tree, n):
                q2.append(i)
        if q2!=[]:
            q1 = q2.copy()
            q2 = []
            level+=1
        elif q2==[]:
            break
    print(co)
    colors = []
    for n in G.nodes:
        colors.append(co.get(n,level+3))
    print(colors)
    nx.draw_networkx(G,node_color=colors, cmap=plt.get_cmap('YlOrRd'))
```

```python
def get_giant_size(n,npv):
    size = []
    for i in range(10):
        G = nx.gnp_random_graph(n,1.0*npv/n)
        scomps = sorted(nx.connected_components(G), key = len, reverse=True)
        size.append(1.0*len(scomps[0])/G.number_of_nodes())
    return size
```

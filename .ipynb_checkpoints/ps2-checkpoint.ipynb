```python
import os
```

```python
G_cycle = nx.cycle_graph(1000)
G_complete = nx.complete_graph(1000)
```

```python
all_methods = [nx.write_graphml
            ,nx.write_pajek
            ,nx.write_gml
            ,nx.write_gexf
            ,nx.write_edgelist
            ,nx.write_adjlist]

all_methods = reversed(all_methods)

for method in all_methods:
    cy_name = f"{method.__name__}-cycle.txt"
    com_name = f"{method.__name__}-complete.txt"
    method(G_cycle, cy_name)
    method(G_complete, com_name)
    cy_size = os.stat(cy_name).st_size /1024
    com_size = os.stat(com_name).st_size /1024
    print(f"{method.__name__[6:]} --- cy:{int(cy_size)}kB - com:{int(com_size)}kB")
    
    
```

```python
# Intersting #1
G_polblog = nx.read_gml("polblogs/polblogs.gml",label='id')
#G_polblog = nx.read_edgelist('/Users/xieewenz/python/Networkx/out.dimacs10-polblogs')
#plt.figure(figsize=(18,18))
#nx.draw_networkx(G_polblog,with_label=false)
#dir(G_polblog)
```

```python
c=sorted(nx.connected_components(G_polblog), key = len, reverse=True)
cl = list(c)
H = G_polblog.subgraph(cl[0])
```

```python
color = []
for node in H.nodes:
    color.append(H.nodes[node]['value'])
```

```python
plt.figure(figsize=(12,12))

nx.draw_networkx(H, with_labels=False, node_size=10, node_color=color)
plt.axis('off')
# nx.draw_networkx(G_polblog, cmap=plt.get_cmap('viridis'), node_color=parties, with_labels=False, font_color='white')
# plt.show()
```

```python
G_fb = H
pos = nx.spring_layout(G_fb)
betCent = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
node_color = [20000.0 * G_fb.degree(v) for v in G_fb]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                 node_color=color,
                 node_size=node_size )
plt.axis('off')
```

```python
G_cycle = nx.cycle_graph(1000)
G_complete = nx.complete_graph(1000)
```

```python
all_methods = [nx.write_graphml
            ,nx.write_pajek
            ,nx.write_gml
            ,nx.write_gexf
            ,nx.write_edgelist
            ,nx.write_adjlist]

all_methods = reversed(all_methods)

for method in all_methods:
    cy_name = f"{method.__name__}-cycle.txt"
    com_name = f"{method.__name__}-complete.txt"
    method(G_cycle, cy_name)
    method(G_complete, com_name)
    cy_size = os.stat(cy_name).st_size /1024
    com_size = os.stat(com_name).st_size /1024
    print(f"{method.__name__[6:]} --- cy:{int(cy_size)}kB - com:{int(com_size)}kB")
    
    
```

#### Helper Functions

```python
def find_giant_component(G):
    # input: G <nx.Graph>
    # output: (giant_component <nx.Graph>, size_of_component <int> )
    c=sorted(nx.connected_components(G), key = len, reverse=True)
    cl = list(c)
    H = G.subgraph(cl[0])
    size = len(cl[0])
    return (H, size)
```
